```python
#import libraries
import numpy as np
import pandas as pd
```

# Chapter 7

# Remove duplicates

```python
print("Before removing duplicates",len(df_boing))
df_boing.drop_duplicates(subset ="Timestamp",keep = False, inplace = True)
print("After removing duplicates",len(df_boing))
```

```python
print("Before removing duplicates",len(df_candy))
df_candy.drop_duplicates(subset ="Timestamp",keep = False, inplace = True)
print("After removing duplicates",len(df_candy))
```

# Replacing Values

```python
df_boing['Which country do you live in?'].unique()
```

```python
df_boing['Which country do you live in?'] = df_boing['Which country do you live in?'].str.upper()
```

```python
df_boing['Which country do you live in?'].unique()
```

```python
df_boing['Which country do you live in?'] = df_boing['Which country do you live in?'].replace(['USA', 'UNITED STATES', 'US','USA ','UNITED STATES OF AMERICA', 'USA USA USA', 'UNITED STATES ', 'UNITED STETES','UNITED  STATES OF AMERICA','THE YOO ESS OF AAAYYYYYY'],'USA')
```

```python
df_boing['Which country do you live in?'].unique()
```

# Filter out missing data

```python
df_boing.dropna(how='all', axis=1, inplace=True)
```

```python
df_boing
```

```python
df_boing = df_boing.dropna(thresh=10)
```

```python
df_boing
```

```python
df_candy.dropna(how='all', axis=1, inplace=True)
```

```python
df_candy
```

```python
df_candy = df_candy.dropna(thresh=10)
```

```python
df_candy
```

# Chapter 8

# Combine and Merge Datasets

```python
df = pd.merge(df_boing,df_candy,on="Timestamp",how="left",suffixes=('_boing','_candy'))
```

```python
df.head()
```

# Pivot the data

```python
df.pivot(index='Timestamp',columns="How old are you?_boing")
```

# Chapter 10

# Grouping with Dicts/Series

```python
d=dict()
for i in range(1,100):
    d[i]=i

df['group']=df['How old are you?_boing'].map(d)
grouped=df.groupby('group')

for group_name, data in grouped:
    print(group_name)
    print('---------')
    print(data)
```

# Grouping with Index Levels

```python
df.set_index(['Timestamp','How old are you?_boing'])
```

# Chapter 11

# Convert between string and date time

```python
df['Timestamp'] = df['Timestamp'].dt.strftime('%Y-%m-%d %H:%M')
```

```python
df.head(10)
```

# Convert timestamps to periods and back

```python
df["Periods"]=pd.DatetimeIndex(df["Timestamp"]).to_period('M')
```
