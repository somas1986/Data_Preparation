# ACTIVITY 7: Reading Tabular Data from a Web Page and Creating DataFrames

```python
#Import libraries
import pandas as pd
from bs4 import BeautifulSoup
from urllib.request import urlopen
import matplotlib.pyplot as plt
```

```python
url='https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(nominal)'
```

# Read page using bs4

```python
html = urlopen(url) 
soup = BeautifulSoup(html, 'html.parser')
```

# Find the table structure you will need to deal with (how many tables there are?)

```python
tables = soup.find_all("table")
```

```python
print(type(tables))
```

```python
print(len(tables))
```

# Find right table using bs4

```python
gdp_table = soup.find("table", { 'class' : 'wikitable' })
```

```python
print(type(gdp_table))
```

```python
print(len(gdp_table))
```

```python
df_table = pd.read_html(str(gdp_table))

df_data = df_table[0]
df_data
```

# Separate the source names and their corresponding data

```python
sources = gdp_table.tbody.findAll('tr', recursive=False)[0]
sources_list = [th for th in sources.findAll('th')]
print(len(sources_list))
```

```python
print(sources_list)
```

```python
data_tables = gdp_table.tbody.findAll('tr')[1:]
```

```python
data_tables
```

```python
len(data_tables)
```

# Get list of source names from the list of sources you have created

```python
source_names = []
for source in sources_list:
    source_name = source.text.strip()
    source_names.append(source_name)
print(source_names[2:])
```

# Seperate the header and data for the first source

```python
header1 = [th.getText().strip() for th in gdp_table.findAll('tr')[0].findAll('th')][:3]
header1
```

```python
data_IMF = df_data[header1]
data_IMF
```

# Do the same for the other two sources

```python
header_all = [th.getText().strip() for th in gdp_table.findAll('tr')[0].findAll('th')]

header2 = []
header2.append(header_all[0])
header2.append(header_all[1])
header2.append(header_all[3])
header2
```

```python
data_UN = df_data[header2]
data_UN
```

```python
header_all = [th.getText().strip() for th in gdp_table.findAll('tr')[0].findAll('th')]

header3 = []
header3.append(header_all[0])
header3.append(header_all[1])
header3.append(header_all[4])
header3
```

```python
data_WB = df_data[header3]
data_WB
```

# ACTIVITY 8: Handling Outliers and Missing Data

# Read the visit_data.csv file

```python
df_visit = pd.read_csv("visit_data.csv")
```

```python
df_visit.head()
```

# Check for duplicates

```python
duplicates = df_visit.duplicated()
print("Duplicate Rows :  " + format(any(duplicates)))
```

```python
duplicates_fn = df_visit.first_name.duplicated()
print("Duplicates for First Name :  " + format(any(duplicates_fn)))
```

```python
duplicates_ln = df_visit.last_name.duplicated()
print("Duplicate for Last Name :  " + format(any(duplicates_ln)))
```

```python
duplicates_email = df_visit.email.duplicated()
print("Duplicate for Email :  " + format(any(duplicates_email)))
```

```python
duplicates_gender = df_visit.gender.duplicated()
print("Duplicate for Gender :  " + format(any(duplicates_gender)))
```

```python
duplicates_ip = df_visit.ip_address.duplicated()
print("Duplicate for IP Address :  " + format(any(duplicates_ip)))
```

```python
duplicates_visit = df_visit.visit.duplicated()
print("Duplicate for Visit :  " + format(any(duplicates_visit)))
```

# Check if any essential column contains NaN

```python
print("First Name column contains NaN: "+format(any(df_visit['first_name'].isnull())))
```

```python
print("Last Name column contains NaN: "+format(any(df_visit['last_name'].isnull())))
```

```python
print("Email column contains NaN: "+format(any(df_visit['email'].isnull())))
```

```python
print("Gender column contains NaN: "+format(any(df_visit['gender'].isnull())))
```

```python
print("IP address column contains NaN: "+format(any(df_visit['ip_address'].isnull())))
```

```python
print("Visit column contains NaN: "+format(any(df_visit['visit'].isnull())))
```

# Get rid of outliers

```python
from scipy import stats
import numpy as np

size_prev = df_visit.shape
df_visit = df_visit[np.isfinite(df_visit['visit'])] #This is an inplace operation. After this operation the original DataFrame is lost.
size_after = df_visit.shape

size_prev
size_after
```

# Report the size difference

```python
size_after
```

```python
size_prev
```

```python
print("Size of the Dataframe with outliers is {}  and without outliers is {}".format(size_prev,size_after)) 
```

# Create a box plot to check for outliers

```python
plt.boxplot(df_visit.visit, notch=True)
```

# Get rid of any outliers

```python
df_visit_outliers = df_visit[(df_visit['visit'] <= 2900) & (df_visit['visit'] >= 100)]
```

```python
print("After getting rid of outliers the new size of the data is - {}".format(*df_visit_outliers.shape))
```
